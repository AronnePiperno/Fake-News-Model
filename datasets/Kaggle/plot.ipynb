```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

df = pd.read_csv('news.csv')

print(df.head())

df['label'] = df['label'].replace(['fake', 'true'], ['FAKE', 'REAL'])

# Define the color palette from green to red
colors = sns.color_palette("RdYlGn_r", len(df['label'].unique()))

# Set the figure size
plt.figure(figsize=(10, 6))

# Plot the count of each label with the defined color palette and label order
ax = sns.countplot(x='label', data=df, palette=colors, order = ['REAL', 'FAKE'])

# Set the title and axis labels
ax.set_title('Distribution of Labels KAGGLE dataset')
ax.set_xlabel('Label')
ax.set_ylabel('Count')

# Rotate x-axis labels for better readability
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

# Add a grid
ax.grid(True, axis='y', linestyle='--')

plt.tight_layout()

# Save the plot
plt.savefig('label_distribution_KAGGLE.png', dpi=300)
```
