```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('WELFake_Dataset.csv')

print(df.head())
```

```python
# change the label 

df['label'] = df['label'].replace([0, 1],['REAL', 'FAKE'] )
```

```python
print(df.head())
```

```python
# Define the color palette from green to red
colors = sns.color_palette("RdYlGn_r", len(df['label'].unique()))

# Set the figure size
plt.figure(figsize=(10, 6))

# Plot the count of each label with the defined color palette and label order
ax = sns.countplot(x='label', data=df, palette=colors, order=['REAL', 'FAKE'])

# Set the title and axis labels
ax.set_title('Distribution of Labels WELFAke dataset')
ax.set_xlabel('Label')
ax.set_ylabel('Count')

# Rotate x-axis labels for better readability
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

# Add a grid
ax.grid(True, axis='y', linestyle='--')

plt.tight_layout()

# Save the plot
plt.savefig('label_distribution_WELFake.png', dpi=300)
```
